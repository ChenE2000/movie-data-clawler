```python
import os
import json
```

```python
# 读取/metadata/merged下的所有json
ls = os.listdir('../metadata/merged')
len(ls)
```

```python
movies = []
for file in ls:
    path = os.path.join('../metadata/merged', file)
    with open(path, 'r', encoding='utf-8') as f:
        movies.append(json.load(f))
        
```

```python
len(movies)
```

```python
# filter id_MaoYan and id_DouBan not None
intact_movies = list(filter(lambda x: x["id_MaoYan"] is not None and x["id_DouBan"] is not None, movies))
len(intact_movies)
```

```python
# save intact_movies to json
with open('../metadata/intact_movies_ids.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(intact_movies, ensure_ascii=False, indent=4))
```

```python
with open("../metadata/movies_boxoffice.json", 'r', encoding='utf-8') as f:
    boxoffice = json.load(f)
```

```python
movies_with_boxoffice = list(filter(lambda x: x['profession_films_boxoffice'] != 0, boxoffice))
```

```python
len(movies_with_boxoffice)
```

```python
with open("../metadata/intact_movies_ids.json", 'r', encoding='utf-8') as f:
    ids = json.load(f)
```

```python
ids
```

```python
# merge ids and movies_with_boxoffice with title
for movie in movies_with_boxoffice:
    title = movie['title']
    # find movie in ids with same title
    _ = list(filter(lambda x: x['title'] == title, ids))
    if len(_) == 0:
        continue
    else:
        movie['ids'] = _[0]
        
```

```python
movies_with_boxoffice
```

```python
# 去除movies_with_boxoffice中没有ids的电影
movies_with_boxoffice = list(filter(lambda x: 'ids' in x, movies_with_boxoffice))
```

```python
movies_with_boxoffice
```

```python
len(movies_with_boxoffice)
```

```python
# save movies_with_boxoffice as json
with open('../metadata/movies_with_boxoffice.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(movies_with_boxoffice, ensure_ascii=False, indent=4))
```
